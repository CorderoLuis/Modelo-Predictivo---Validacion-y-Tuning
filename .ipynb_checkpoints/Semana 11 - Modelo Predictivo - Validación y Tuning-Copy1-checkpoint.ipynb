## Modelo Predictivo - Validación Y Tuning

Usando los datos de **autocasion_procesado.csv** vamos a realizar un **GridSearchCV()** para encontrar los mejores hiperparámetros para el mejor modelo de la semana anterior.

### Preprocesamiento

- Para este ejercicio vamos a usar el **30%** de filas aleatorias del total del **DataFrame**, haremos esto para evitar tiempo de calculos muy grandes. El código de esto ya está implementado. Usaremos _**random_state = 42**_
- Puedes hacer **Feature Selection** para elegir las mejores columnas. Haz esto después de filtrar los datos.
- Recuerda escalar los datos antes de hacer el **GridSearchCV()**.
- Define **X** e **y**. Donde **y** es la columna _**precio**_ y **X** es el resto de columnas que hayas selccionado usando **Feature Selection**.

### GridSearchCV()

Realiza un **GridSearchCV()** usando **RandomForestRegressor()**

- Define un **diccionario** con los parámetros a utilizar del modelo **RandomForestRegressor()**.
- Define un **set** con las métricas a evaluar, usa _**mean_absolute_error**_, _**mean_squared_error**_ y _**r2_score**_.
- Usa el parámetro _**cv = 5**_ en **GridSearchCV()**.
- La métrica a monitorear será _**r2_score**_.

### Resultados

Una vez encontrado los parámetros para el mejor modelo:

- Imprime por pantalla los parámetros encontrados.
- Entrena un nuevo modelo de **RandomForestRegressor()** usando estos parámetros, usa **X** e **y** para entrenar.
- Guarda el modelo en un archivo **.pkl**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Escaladores
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# Train, Test
from sklearn.model_selection import train_test_split

# Modelos
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

# Métricas
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from sklearn.model_selection import GridSearchCV

import time
```

```python
df = pd.read_csv("autocasion_procesado.csv")

df = df.sample(frac = 0.3, random_state = 42, ignore_index = True)

df.shape
```

```python
df
```

```python
X = df.drop(["precio"], axis = 1)
y = df["precio"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape},  y_test: {y_test.shape}")
```

```python
%%time

model = RandomForestRegressor(n_estimators=250, random_state=0, n_jobs=-1)

# Entrenamiento
model.fit(X, y)

# Importancia de las características
importances = model.feature_importances_

# DataFrame con las importancias
df_importances = pd.DataFrame(data=zip(X.columns, importances), columns=["Columnas", "Importancia"])

# Ordenamos el DataFrame por importancia
df_importances = df_importances.sort_values("Importancia", ascending=False)


print("Feature Importance:")
for index, (feature, importance) in enumerate(df_importances.values):
    print(f"{index + 1:2}. feature {index + 1:2} ({importance:20}): {feature}")

# Grafica importancia
plt.figure(figsize=(12, 8))
plt.title("Feature Importances")
sns.barplot(x=df_importances["Importancia"], y=df_importances["Columnas"],color ='blue')
plt.grid()
plt.show()
```

```python
X = np.array(df[['Potencia (cv)','Kilómetros','year','peso_masa_kg']])
y = np.array(df['precio'])
X.shape,y.shape
```

```python
X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()
```

```python
X_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_scaler.fit(y_train.to_numpy().reshape(-1, 1))
y_train_scaled = y_scaler.transform(y_train.to_numpy().reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.to_numpy().reshape(-1, 1))
```

```python
#Definir modelo y Grid_search
```

```python
%%time

model = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [5, 10, 15 , None],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [2, 4, 6],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['squared_error']
}


scores = {
    'mae': 'neg_mean_absolute_error', 
    'mse': 'neg_mean_squared_error', 
    'r2': 'r2'
}

grid_solver = GridSearchCV(estimator=model, 
                           param_grid=param_grid, 
                           scoring =scores, 
                           cv=5, 
                           refit='r2', 
                           n_jobs=-1, 
                           verbose=3)

model_result = grid_solver.fit(X_train_scaled, y_train)

print(model_result.best_score_)
print(model_result.best_params_)
```

```python
%%time

model = AdaBoostRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1],
    'loss': ['linear', 'square', 'exponential']
}

scores = {
    'mae': 'neg_mean_absolute_error', 
    'mse': 'neg_mean_squared_error', 
    'r2': 'r2'
}
    
    
grid_solver = GridSearchCV(estimator=model, 
                           param_grid=param_grid, 
                           scoring =scores, 
                           cv=5, 
                           refit='r2', 
                           n_jobs=-1, 
                           verbose=3)  

model_result = grid_solver.fit(X_train_scaled, y_train)

print(model_result.best_score_)
print(model_result.best_params_)
```
